```python
import scipy.io
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Load .mat file
mat_data = scipy.io.loadmat('Cluster_dataAllPat.mat')
clusters = mat_data['Cluster_dataAllPat'].squeeze()

# Drop Cluster 11 (index 10 in Python)
clusters = np.delete(clusters, 10)
cluster_names = [f'Cluster {i+1}' for i in range(15)]

# ExcelWriter to save per-cluster gene expression data and top 10 genes summary
cluster_data_writer = pd.ExcelWriter("cluster_gene_data_foldchange.xlsx", engine='xlsxwriter')

# Function to perform PCA and return top genes
def perform_pca_with_foldchange(cluster_data, cluster_name, n_top_genes=10):
    gene_names = []
    expr_matrix = []

    for row in cluster_data:
        gene_name = row[0].item() if isinstance(row[0], np.ndarray) else str(row[0])
        gene_expr = np.array(row[1:]).squeeze().astype(np.float64)
        gene_names.append(gene_name)
        expr_matrix.append(gene_expr)

    gene_names = np.array(gene_names)
    expr_matrix = np.stack(expr_matrix)  # (n_genes, n_timepoints)

    # Compute fold change w.r.t. first timepoint and apply log2(1 + FC)
    fc_matrix = expr_matrix / expr_matrix[:, [0]]
    log_fc_matrix = np.log2(1 + fc_matrix)
    log_fc_matrix = np.nan_to_num(log_fc_matrix)

    # Save processed data to Excel
    timepoint_cols = [f"TP{i+1}" for i in range(log_fc_matrix.shape[1])]
    df_cluster = pd.DataFrame(log_fc_matrix, index=gene_names, columns=timepoint_cols)
    df_cluster.reset_index(inplace=True)
    df_cluster.rename(columns={'index': 'Gene'}, inplace=True)
    df_cluster.to_excel(cluster_data_writer, sheet_name=cluster_name.replace(" ", "_"), index=False)

    # PCA on log2 fold change matrix
    pca = PCA(n_components=2)
    pca.fit(log_fc_matrix)

    # Gene contributions to PC1
    gene_contributions = log_fc_matrix @ pca.components_[0]
    gene_contrib_series = pd.Series(np.abs(gene_contributions), index=gene_names)
    top_genes = gene_contrib_series.sort_values(ascending=False).head(n_top_genes)

    # Print top drivers
    print(f"\nTop {n_top_genes} driver genes for {cluster_name}:")
    for gene, score in top_genes.items():
        print(f"{gene}: {score:.4f}")

    # Plot PCA
    scores = pca.transform(log_fc_matrix)
    plt.figure(figsize=(7, 6))
    sns.scatterplot(x=scores[:, 0], y=scores[:, 1], s=30, color='gray')

    for gene in top_genes.index:
        idx = np.where(gene_names == gene)[0][0]
        plt.text(scores[idx, 0], scores[idx, 1], gene, fontsize=9, weight='bold')

    plt.title(f"PCA (log2 Fold Change) - {cluster_name}")
    plt.xlabel("PC1")
    plt.ylabel("PC2")
    plt.tight_layout()
    plt.show()

    # Return top genes as a DataFrame
    df_top_genes = pd.DataFrame({
        'Gene': top_genes.index,
        'PC1_Contribution': top_genes.values,
        'Cluster': cluster_name
    })

    return df_top_genes

# Collect all top genes from all clusters
top_genes_all_clusters = pd.DataFrame()

# Run PCA for all clusters
for i, cluster_data in enumerate(clusters):
    cluster_name = cluster_names[i]
    df_top_genes = perform_pca_with_foldchange(cluster_data, cluster_name)
    top_genes_all_clusters = pd.concat([top_genes_all_clusters, df_top_genes], ignore_index=True)

# Save top genes summary to Excel
top_genes_all_clusters.to_excel(cluster_data_writer, sheet_name="Top10_Genes_All_Clusters", index=False)

# Save Excel file
cluster_data_writer.close()
```
